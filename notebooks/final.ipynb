```python
# Load data from gzipped JSON lines file
def load_jsonl_gz(filepath):
    data = []
    with gzip.open(filepath, 'rt', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return data

# Load training data
training_data = load_jsonl_gz(train_data_path)

# Prepare dictionaries from training data
username2posts_train = {entry["username"]: entry["posts"] for entry in training_data}
username2_category = {entry["username"]: entry["category"] for entry in training_data}

# Placeholder for testing data (load as needed for classification or regression tasks)
username2posts_test = {}  # Populate based on your specific testing requirements
```

```python
import re

def preprocess_text(text):
    """Preprocess text by cleaning and normalizing."""
    text = text.casefold()  # Lowercase
    text = re.sub(r'http\S+|www\S+', '', text)  # Remove URLs
    text = re.sub(r'#\w+', 'HASHTAG', text)  # Replace hashtags
    text = re.sub(r'@\w+', 'MENTION', text)  # Replace mentions
    text = re.sub(r'[^a-zçğıöşü0-9\s]', '', text)  # Keep specific characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
    return text
```
